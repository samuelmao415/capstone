# Products

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('all.csv').drop('Unnamed: 0', axis=1).drop_duplicates('alltext')
```

```python
df.groupby('category')['id'].count()
```

```python
import re
df=df.dropna(subset=['price'])
df.price = df.price.map(lambda x: float(re.sub('[$,]','', x)))
df.shape
```

```python
df.groupby(['website'])['price'].mean().plot.bar()
```

# Instagram

```python
insta = pd.read_csv("insta.csv").drop("Unnamed: 0", axis=1)
```

```python
insta.head()
```

```python
insta.groupby('blogger')['likes'].mean().reset_index().sort_values('likes').plot.bar(x='blogger')

# kylie jenenr has most likes 
```

```python
insta.groupby('blogger')['comments'].mean().reset_index().sort_values('comments').plot.bar(x='blogger')

# kylie jenner has most comments
```

```python
insta.loc[:,'cap_len'] = insta['pic_cap'].apply(lambda x: len(x))
insta
```

```python
insta.groupby('blogger')['cap_len'].mean().reset_index().sort_values('cap_len').plot.bar(x='blogger')

# kylier jenenr has least caption
```
